```python
# Import necessary libraries
import networkx as nx  # NetworkX is a Python library used to work with graphs
import matplotlib.pyplot as plt  # Matplotlib is a plotting library to visualize graphs
from collections import deque  # deque is used to efficiently implement queues and stacks
```

```python
# Function to perform DFS (Depth-First Search) traversal
def dfs(graph, start):
    visited = set()  # A set to keep track of visited nodes to avoid revisiting them
    stack = [start]  # Initialize a stack with the start node. A list is used as a stack.
    visited.add(start)  # Mark the start node as visited

    # Print the DFS traversal order heading
    print("DFS Traversal Order:", end=" ")

    # Continue the DFS process until the stack is empty
    while stack:  # While the stack is not empty
        node = stack.pop()  # Pop a node from the stack (LIFO order)
        print(node, end=" ")  # Print the current node being processed

        # Enqueue all unvisited neighbors of the current node
        for neighbor in graph[node]:  # Iterate through each neighbor of the current node
            if neighbor not in visited:  # If the neighbor hasn't been visited
                visited.add(neighbor)  # Mark the neighbor as visited
                stack.append(neighbor)  # Push the neighbor onto the stack for future exploration
```

```python
# Example graph represented as an adjacency list (dictionary)
graph = {
    1: [2, 3],  # Node 1 is connected to nodes 2 and 3
    2: [1, 4, 5],  # Node 2 is connected to nodes 1, 4, and 5
    3: [1, 6, 7],  # Node 3 is connected to nodes 1, 6, and 7
    4: [2],  # Node 4 is connected to node 2
    5: [2],  # Node 5 is connected to node 2
    6: [3],  # Node 6 is connected to node 3
    7: [3]   # Node 7 is connected to node 3
}
```

```python
# Create a graph object using NetworkX
G = nx.Graph()  # Create an empty undirected graph
```

```python
# Convert the adjacency list into edges and add them to the NetworkX graph object
# This line generates all the edges from the adjacency list and adds them to the graph
edges = [(node, neighbor) for node, neighbors in graph.items() for neighbor in neighbors]
G.add_edges_from(edges)  # Add all the edges to the NetworkX graph
```

```python
# Generate a layout for the graph using the spring layout algorithm
# The spring_layout arranges nodes such that the edges are evenly spaced and the graph is visually appealing
pos = nx.spring_layout(G)
```

```python
# Create a figure and axes for visualization using Matplotlib
# fig, ax = plt.subplots(figsize=(8, 6)) creates a figure object `fig` and axes object `ax`
fig, ax = plt.subplots(figsize=(8, 6))  # Create a figure with a size of 8x6 inches

# Draw the graph using the `nx.draw()` function
# `pos` determines the node positions, `with_labels=True` ensures node labels are shown
# `node_color='skyblue'` sets the color of the nodes to sky blue
# `font_weight='bold'` makes the node labels bold
# `font_size=10` sets the font size of the labels
# `node_size=2000` sets the size of the nodes
# `edge_color='gray'` sets the color of the edges to gray
nx.draw(G, pos, with_labels=True, node_color='skyblue', font_weight='bold', font_size=10, node_size=2000, edge_color='gray', ax=ax)

# Set a title for the plot to describe the content
ax.set_title("Graph Visualization")  # The title of the graph will be displayed at the top

# Display the plot
# This command will render the graph and display it in a separate window
plt.show()
```
